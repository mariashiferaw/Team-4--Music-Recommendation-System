```python
#Import libraries
import pandas as pd
import numpy as np
```

```python
# Reading a csv
sp1 = pd.read_csv('spotify_songs.csv')
```

```python
#DataFrame information
sp1.info()
```

```python
# Check the data for columns with missing numbers (i.e. np.NaN).
sp1.isnull().sum()
```

```python
#copy a sp1
sp2 = sp1.copy()
```

```python
# Display the columns with missing numbers
sp2[sp2['track_name'].isna() == True]
sp2[sp2['track_artist'].isna() == True]
sp2[sp2['track_album_name'].isna() == True]
```

```python
# Drop null value
sp2.dropna(inplace=True)
```

```python
# Check the data for columns with missing numbers (i.e. np.NaN).
sp2.isnull().sum()
```

```python
# Dropping a column
sp2.drop(['track_id'], axis = 1, inplace = True)
sp2.drop(['track_album_id'], axis = 1, inplace = True)
sp2.drop(['playlist_id'], axis = 1, inplace = True)
sp2.drop(['playlist_name'], axis = 1, inplace = True)
sp2.drop(['playlist_subgenre'], axis = 1, inplace = True)
```

```python
# Rename column
sp2.rename({'track_name':'title',\
              'track_artist':'artist',\
              'track_popularity':'popularity',\
              'track_album_name':'album',\
              'track_album_release_date':'release_date',\
              'playlist_genre':'genre'
            }, axis = 1, inplace = True)
```

```python
# Let's convert duration to minute 
sp2["duration_min"] = (sp2["duration_ms"] * 1)/60000
sp2["duration_min"]
```
