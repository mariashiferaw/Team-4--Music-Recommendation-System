```python
df1 = df.dropna()

df1
```

```python
#Check for Duplicated Values
df1.duplicated().sum()
```

```python
#Drop unnecessary columns 
spotify_songs = df1.drop(columns=['Unnamed: 0','track_album_id','track_album_release_date','key','mode'])

spotify_songs
```

```python
#Convert duration from milliseconds to seconds 
spotify_songs['duration_ms'] = spotify_songs['duration_ms']*1000

# Renaming the column
spotify_songs.rename(columns={'duration_ms':'duration_s'}, inplace=True)

spotify_songs.head()
```

```python
spotify_songs.info()
```

```python
features = spotify_songs[['danceability', 'energy', 'loudness','speechiness','acousticness','instrumentalness','liveness','tempo','valence','duration_s']]                  
```

```python
# Feature Correlation Matrix
import seaborn as sns
import matplotlib.pyplot as plt

music_features = ['danceability', 'energy', 'loudness','speechiness','acousticness','instrumentalness','liveness','tempo','valence','duration_s']
#Visualizing outliners by using box plots
plt.figure(figsize=(12,10))

for i,feature in enumerate(music_features,1):
    plt.subplot(5,2,i)
    sns.boxplot(y=spotify_songs[feature])
    plt.title(feature)

plt.show()
```

```python
#Using Interquartile Range  (IQR) to identify outliers in a data set.

Q1 = spotify_songs[music_features].quantile(0.25)
Q3 = spotify_songs[music_features].quantile(0.75)

IQR = Q3 - Q1

outliers = ((spotify_songs[music_features] < (Q1 - 1.5 * IQR)) | (spotify_songs[music_features] > (Q3 + 1.5 * IQR))).any(axis=1)
outliers_indices = spotify_songs[outliers].index
num_outlier = outliers.sum()

num_outlier,outliers_indices

```

```python
#Calculate the correlation matrix 
corr_matrix = features.corr()

#Create heatmap to visualize the correlation matrix
plt.figure(figsize=(10,10))
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm',fmt='.2f')
plt.title('Feature Correlation Matrix')
plt.show()
```

```python
#Analyzing 

#Danceability and Valence have a positive correletion (0.8). More danceable often have more postive mood
#Energy and Loudness also show a strong positive correlation. Louder tracks are often perceived as more energetic 
#Acousticness shows a negative correlation with energy and loudness. More acoustic tracks tend to be quier and less energic.
#Speechiness has moderate positive correlation with danceability and valence 
#Tempo also has a moderte positive with energy. faster-paced songs tend to be more energetic.
#Liveness shows a weak to moderate correlation with energy and loudness. Live recordings might have slightly higher energy and loudness, not significantly. 
```
