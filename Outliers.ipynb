```python
import pandas as pd
```

```python
data = pd.read_csv('cleaned_data.csv')
data.head()
```

```python
data.shape
```

```python
features = ['danceability', 'energy','key','loudness','mode','speechiness','acousticness','instrumentalness','liveness','tempo','valence','duration_ms']                
```

```python
Q1 = data[features].quantile(0.25)
Q3 = data[features].quantile(0.75)

IQR = Q3 - Q1

outliers = ((data[features] < (Q1 - 1.5 * IQR)) | (data[features] > (Q3 + 1.5 * IQR))).any(axis=1)
outliers_indices = data[outliers].index
num_outlier = outliers.sum()

num_outlier,outliers_indices
```

```python
outliers = {}
for feature in features:
    Q1 = data[feature].quantile(0.25)
    Q3 = data[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers[feature] = data[(data[feature] < lower_bound) | (data[feature] > upper_bound)]
```

```python
print(outliers)
```

```python
# Print the number of outliers for each feature
print("Number of outliers for each feature:")
for feature, outlier_data in outliers.items():
    print(f"{feature}: {len(outlier_data)}")
```
